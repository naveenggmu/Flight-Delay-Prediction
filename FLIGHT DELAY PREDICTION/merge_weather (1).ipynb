```python
import pandas as pd 
import json
import os
from pprint import pprint
import datetime
```

```python
#os.listdir('Weather/ATL/')
weather_data=[]
weather_data1=[]

for i in os.listdir('Weather'):
    print(i)
    for j in os.listdir('Weather/'+i):
        if('2016' in j):
            print(j)
            temp=json.load(open('Weather/'+i+'/'+j))
            for k in temp['data']['weather']:
#                 print(len(k))
                tempdata=pd.DataFrame(k['hourly'],columns=['windspeedKmph','weatherCode','precipMM','WindGustKmph','pressure','cloudcover','winddirDegree','humidity','cloudcover','DewPointF','tempF','time','WindChillF'])
#                 print(tempdata.shape)
                tempdata['date']=k['date']
                tempdata['place']=i 
                weather_data.append(tempdata)
        if('2017' in j):
            print(j)
            temp1=json.load(open('Weather/'+i+'/'+j))
            for q in temp1['data']['weather']:
#                 print(len(k))
                tempdata1=pd.DataFrame(q['hourly'],columns=['windspeedKmph','weatherCode','precipMM','WindGustKmph','pressure','cloudcover','winddirDegree','humidity','cloudcover','DewPointF','tempF','time','WindChillF'])
#                 print(tempdata1.shape)
                tempdata1['date']=q['date']
                tempdata1['place']=i 
                weather_data1.append(tempdata1)
        
```

```python
weather16df=pd.concat(weather_data,axis=0)
weather17df=pd.concat(weather_data1,axis=0)
```

```python
weather16df.to_csv('weather16.csv')
weather17df.to_csv('weather17.csv')
```
