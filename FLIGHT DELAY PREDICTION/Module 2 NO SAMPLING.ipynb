```python
X=merge.drop(["DepTime","DepDel15","DepDelayMinutes","ArrDelayMinutes","ArrTime","ArrDel15","Unnamed: 0","Unnamed: 0_x","Unnamed: 0_y","Unnamed: 0.1","Origin","Dest","airport_x","airport_y","FlightDate","ArrFlightDate","date_y","date_x","time_x","time_y","DepAbsTime","ArrAbsTime"],axis=1)
X
```

```python
X.transpose()
```

```python
X.shape
```

```python
Y=merge["ArrDel15"]
```

```python
import collections
collections.Counter(Y)
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 42) 
```

```python
#EXTRATREESCLASSIFIER

from sklearn.ensemble import ExtraTreesClassifier
etc=ExtraTreesClassifier(n_jobs=-1,random_state = 42)
etc.fit(X_train,Y_train)
```

```python
Y_predetc=etc.predict(X_test)
```

```python
#ACCURACY
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_predetc)
```

```python
from sklearn import metrics
cm=metrics.confusion_matrix(Y_test,Y_predetc)
print(cm)
```

```python
from sklearn.metrics import classification_report
target_names=["Class 0","Class 1"]
print(classification_report(Y_test,Y_predetc,target_names=target_names))
```

```python
#DECISION TREE
from sklearn.tree import DecisionTreeClassifier

clf=DecisionTreeClassifier()
clf.fit(X_train, Y_train)
```

```python
Y_preddec=clf.predict(X_test)
```

```python
#from sklearn import metrics
cm=metrics.confusion_matrix(Y_test,Y_preddec)
print(cm)
```

```python
#from sklearn.metrics import classification_report
target_names=["Class 0","Class 1"]
print(classification_report(Y_test,Y_preddec,target_names=target_names))
```

```python
#Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,Y_train)
```

```python
Y_predgau=gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_predgau)
```

```python
#from sklearn import metrics
cm=metrics.confusion_matrix(Y_test,Y_predgau)
print(cm)
```

```python
#from sklearn.metrics import classification_report
target_names=["Class 0","Class 1"]
print(classification_report(Y_test,Y_predgau,target_names=target_names))
```
