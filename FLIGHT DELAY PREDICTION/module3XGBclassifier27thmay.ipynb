```python
import pandas as pd
import os
import numpy as np  
import matplotlib.pyplot as plt
```

```python
os.chdir("D:\ML\merge")
```

```python
merge=pd.read_csv("module2resultXGB.csv")
merge.transpose()
```

```python
X = merge.copy()
```

```python
X=X.loc[(X['NewArr15']==1)]
```

```python
X.shape
```

```python
Y = X['ArrDelayMinutes']
```

```python
X.shape
```

```python
Y.shape
```

```python
X=X.drop(['ArrDelayMinutes','Unnamed: 0'],axis = 1)
```

```python
X.transpose()
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20)  
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
##linear regression
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train)
```

```python
Y_pred_lin_test=reg.predict(X_test)
```

```python
print("RMSE  " , np.sqrt(mean_squared_error(Y_test, Y_pred_lin_test)))
print("MAE   " , mean_absolute_error(Y_test, Y_pred_lin_test))
print("R2    " , r2_score(Y_test,Y_pred_lin_test))
```

```python
## EXTRATREESREGRESSOR Tree
from sklearn.ensemble import ExtraTreesRegressor
etr=ExtraTreesRegressor()
etr.fit(X_train,Y_train)
```

```python
Y_pred_etr_test = etr.predict(X_test)
```

```python
print("RMSE  " , np.sqrt(mean_squared_error(Y_test, Y_pred_etr_test)))
print("MAE   " , mean_absolute_error(Y_test, Y_pred_etr_test))
print("R2    " , r2_score(Y_test,Y_pred_etr_test))
```

```python
#XGB Regressor
import xgboost as xgb
xgbmodel = xgb.XGBRegressor()
xgbmodel.fit(X_train,Y_train)
```

```python
Y_pred_xgbr = xgbmodel.predict(X_test)
```

```python
print("RMSE  " , np.sqrt(mean_squared_error(Y_test, Y_pred_xgbr)))
print("MAE   " , mean_absolute_error(Y_test, Y_pred_xgbr))
print("R2    " , r2_score(Y_test,Y_pred_xgbr))
```
