```python
import json
import pandas as pd
from pprint import pprint
path=r'D:\ML\weather\ATL\2016-1.json'
json_string=open(path).read() 
json_object=json.loads(json_string)
pprint(json_object)
    
```

```python
with open(path) as f:
    d = json.load(f)
pprint(d) 
```

```python
df = pd.io.json.json_normalize(d["data"]["weather"])
df['NEWINDEX']=0
j=0
for i in range(0,31):
    df['NEWINDEX'][i]=j
    j+=1
 

```

```python
dair = pd.io.json.json_normalize(d["data"]["request"])
dair.drop(["type"], axis = 1, inplace = True) 

dair["query"]
```

```python
#df['airport']=dair["query"]
for i in range(0,31):
    df['airport'][i]=dair["query"]
df    
```

```python
df.drop(["astronomy","hourly","maxtempC","maxtempF","mintempC","mintempF","sunHour","totalSnow_cm","uvIndex"], axis = 1, inplace = True)
df
```

```python
df1 = pd.io.json.json_normalize(d["data"]["weather"][0]["hourly"])
df2=pd.io.json.json_normalize(d["data"]["weather"][1]["hourly"])
df2=df2.append(df1)
for x in range(2,31):
    df1 = pd.io.json.json_normalize(d["data"]["weather"][x]["hourly"])
    df2=df2.append(df1)

df2
```

```python
df2["NEWINDEX"]=0
j=0
for i in range(0,744,24):
    df2['NEWINDEX'][i:i+24]=j
    j+=1
df2    
```

```python
result = pd.merge(df,df2,
                 on='NEWINDEX')

result.transpose()
```

```python
result.drop(["NEWINDEX","DewPointC","FeelsLikeF","FeelsLikeC","HeatIndexC","HeatIndexF","WindChillC","WindGustMiles","tempC","weatherDesc","weatherIconUrl","winddir16Point","windspeedMiles"], axis = 1, inplace = True)
result
```
